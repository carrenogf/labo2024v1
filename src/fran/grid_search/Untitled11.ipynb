```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Austral - Ciencia de datos/Laboratorio 1/repo/labo2024v1/src/fran/exp/HT2020/gridsearch.txt",
                 sep="\t")
df
```

```python
df["vminbucket"] = df["minbucket"]*df["min_split"]
df.head()
```
